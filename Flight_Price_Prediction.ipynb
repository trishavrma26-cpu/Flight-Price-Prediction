### Encoding Decisions

| Column | Ordinal/Nominal | Encoding Method |
|---------|----------------|-----------------|
| airline |    nominal            | One-Hot Encoding                |
| source_city |   nominal         |  One-Hot Encoding               |
| departure_time |   nominal      |  One-Hot Encoding                |
| stops |      ordinal            | Label Encoding                |
| arrival_time |   nominal         | One-Hot Encoding                 |
| destination_city |  nominal     | One-Hot Encoding                |
| class |  ordinal                |   Label Encoding              |

```python
df['stops'].unique()
```

```python
df['class'].unique()
```

```python
class_mapping = {
    'Economy': 0,
    'Business': 1
}
```

```python
df['class'] = df['class'].map(class_mapping)
```

```python
stops_mapping = {
    'zero': 0,
    'one': 1,
    'two_or_more': 2
}
```

```python
df['stops'] = df['stops'].map(stops_mapping)
```

```python
df[['class', 'stops']].head()
```

```python
df.info()
```

```python
categorical_cols = [
    'airline',
    'source_city',
    'departure_time',
    'arrival_time',
    'destination_city'
]
```

```python
df = pd.get_dummies(
        df,
        columns=categorical_cols,
        drop_first=True
     )
```
